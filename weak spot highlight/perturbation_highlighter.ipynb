### Notebook to create Highlighted Adversarial Images

```python
from PIL import Image
import numpy as np
```

```python
def create_hline(img_in, y, thickness):

    spot = np.full((160, 105), False, dtype=bool)
    for i in range(y, y+thickness, 1):
        spot[i, :] = True

    sample = img_in.copy()

    sample[spot, 0] = 255
    sample[spot, 1] = 0
    sample[spot, 2] = 0

    return sample   # shape: (160, 105, 3)
```

```python
# Parameters    --- EDIT HERE ---
input_filename = '0_DSC02227.JPG'
output_filename = 'hline-0'
pos_start = 20
pos_to = 50
```

```python
im  = Image.open(f'input_images/{input_filename}')
img = np.array(im) # convert PIL -> numpy
display(im)
```

```python
adversarial = create_hline(img, pos_start, pos_to)
adversarial_image = Image.fromarray(adversarial)
display(adversarial_image)
```

```python
# Save the highlighted adversarial image
adversarial_image.save(
    f'output_images/highlighted-{output_filename}.png')
```
